MATPLOTLIB

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

SCATTER PLOT

```python
a= np.arange(0,100,2)
b=np.arange(0,150,3)
```

```python
plt.scatter(a,b,c="r")
plt.xlabel("a value")
plt.ylabel("b value")
plt.title("relationship between a and b")
plt.savefig('Test.png')
```

```python
plt.plot(a, b, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
```

```python
plt.subplot(2,2,1)
plt.plot(a,b,c="r")
plt.subplot(2,2,2)
plt.plot(a,b, c="y")
plt.subplot(2,2,3)
plt.plot(a,b,c='g')
plt.subplot(2,2,4)
plt.plot(a,b,c='r')
```

```python
x=np.arange(2,40)
y=3 * x+7
plt.plot(x,y,c="g")
plt.xlabel("x value")
plt.ylabel("y value")
plt.title(" x and y plot")
```

BAR PLOT

```python
x=[3, 4, 5 ,8]
y=[6,8,9,3]

x1=[9,6,8,8]
x2=[9,5,6,4]
plt.bar(x, y)
plt.bar(x1, x2)
plt.title("bar plot")
plt.xlabel("x axis ")
plt.ylabel("y axis")
```

HISTOGRAM

```python
x= np.arange(1, 40, 5)
plt.hist(x, align='mid',orientation='vertical')
plt.title("histogram")
plt.show()
```

BOX PLOT

```python
data=np.random.rand(100)
plt.boxplot(data,vert=True,patch_artist=False)
```

pie chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
#explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
